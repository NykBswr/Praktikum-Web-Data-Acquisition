### 2. Crawling website unair news dengan mengkombinasikan Python Request, beautifulsoup, dan for loop.

##### Mencari semua link yang berada di tag < a >

```python
for i in soup.find_all("a"):
    print(i.get("href"))
```

Penjelasan:

Dengan *for* i yang berada di dalam variabel *soup* dan menggunakan fungsi **find_all** untuk menemukan semua **<a>** lalu kita *print* semua i yang berisikan semua **href** menggunakan fungsi **get("href")** untuk mendapatkan **href** di dalam **<a>**.

##### Mencari semua text yang berada di tag < h2 >

```python
for i in soup.find_all("h2"):
    print(i.get_text())
```

Penjelasan:
